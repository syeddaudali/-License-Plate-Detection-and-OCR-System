```python
import os
from ultralytics import YOLO
import cv2
import easyocr
import matplotlib.pyplot as plt
from statistics import mode

# Load pre-trained YOLO model (fine-tune with custom dataset if needed)
model = YOLO('license_plate_detector.pt')

# Specify the folder containing videos
video_folder_path = 'videos'  # Replace with your folder path

# Get all video files from the folder (assuming file extensions .mp4, .avi, etc.)
video_paths = 'E:\\Knowledge Streams\\Mini-Project__CV\\videoscar 45-1.mp4' #[os.path.join(video_folder_path, f) for f in os.listdir(video_folder_path) if f.endswith(('.mp4', '.avi', '.mkv'))]

# Initialize the EasyOCR reader
reader = easyocr.Reader(['en'])
preds = []

# Define a function to display a frame in Jupyter
def show_frame(frame):
    plt.figure(figsize=(10, 10))
    plt.imshow(cv2.cvtColor(frame, cv2.COLOR_BGR2RGB))
    plt.axis('off')
    plt.show()

# Process each video
for video_path in video_paths:
    print(f"Processing video: {video_path}")

    # Read the video
    video = cv2.VideoCapture(video_path)

    while True:
        ret, frame = video.read()
        if not ret:
            break  # Exit the loop when video ends

        # Run YOLO model on the current frame
        results = model(frame)

        # Process YOLO results
        for result in results[0].boxes.data.tolist():
            x1, y1, x2, y2, score, class_id = result

            # Draw a rectangle around the detected object (car/license plate)
            cv2.rectangle(frame, (int(x1), int(y1)), (int(x2), int(y2)), (0, 255, 0), 2)

            # Crop the detected area (license plate)
            cr_img = frame[int(y1):int(y2), int(x1):int(x2)]

            # Perform OCR on the cropped image
            result = reader.readtext(cr_img)

            # Extract OCR results based on confidence and text length
            ocr_res = []
            for k in range(len(result)):
                if (result[k][2]) > 0.3 and len(result[k][1]) > 2:
                    ocr_res.append(result[k][1])
            ocr_res = ''.join(ocr_res)

            print(result, ocr_res, '>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>')
            
            # Update predictions if OCR result is valid
            if len(ocr_res) > 1:
                preds.append(ocr_res.upper())
                res = mode(preds)

            # Display the OCR result on the frame
            cv2.putText(frame, res, (int(x1), int(y1) - 10), cv2.FONT_HERSHEY_SIMPLEX, 0.9, (0, 0, 255), 2)

        # Display the frame with bounding boxes and OCR results in Jupyter
        show_frame(frame)

    # Release the video object
    video.release()
    
show_frame()    

# # Close any OpenCV windows (not necessary for Jupyter)
# cv2.destroyAllWindows()
```
